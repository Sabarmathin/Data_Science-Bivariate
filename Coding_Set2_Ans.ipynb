```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stat
```

```python
#Question 1:
delays = [15, 30, 45, 20, 25, 100, 5, 60, 35, 50, 
          120, 80, 10, 75, 90, 200, 55, 40, 70, 150]

delay_percen = np.percentile(delays,[10,25,50,75,90])
print("Percentiles (10th, 25th, 50th, 75th, 90th): ", delay_percen)
Q1,Q3 = np.percentile(delays,[25,75])
IQR = Q3-Q1
print("IQR:",IQR)
outlier = [x for x in delays if (x < (Q1-IQR*(1.5))) or (x > (Q3+IQR*(1.5)))]
print("Outliers:",outlier)
plt.xlabel('Flight Delays (minutes)')
plt.title('Box Plot for Flight Delays')
sns.boxplot(x = delays)
plt.show()
```

```python
#Question 2:
salaries = [30000, 32000, 35000, 37000, 40000, 42000, 45000, 47000, 50000, 55000,
            60000, 62000, 65000, 67000, 70000, 72000, 75000, 80000, 85000, 90000]
Mean = np.mean(salaries)
print("Mean Salary:",Mean) 
Median = np.median(salaries)
print("Median Salary:",Median)
Mode = stat.mode(salaries)
print("Mode Salary:",Mode)
skew = stat.skew(salaries)
print("The data is positive skeew:",skew)
sns.histplot(salaries,kde = True)
plt.title('Histogram of Employee Salaries')
plt.show()
```

```python
#Question 3:
data = {
    'Student': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Mathematics': [85, 78, 92, 88, 70, 95, 60, 80, 90, 76],
    'Science': [80, 85, 88, 70, 75, 92, 55, 82, 89, 78],
    'English': [78, 74, 85, 80, 68, 90, 50, 77, 83, 72]
}
df_marks = pd.DataFrame(data)
print("Descriptive Statistics \n",df_marks.describe())
sns.histplot(df_marks,kde = True)
plt.title('Histogram(graph)s')
plt.show()
sns.histplot(df_marks['Mathematics'],kde = False)
plt.title('Histogram graph - Mathematics')
plt.show()
sns.histplot(df_marks['Science'],kde = False)
plt.title('Histogram graph - Science')
plt.show()
sns.histplot(df_marks['English'],kde = False)
plt.title('Histogram graph - English')
plt.show()
correlation = df_marks.corr(numeric_only=True)
print("Correlation Analysis:\n",correlation)
# sns.heatmap(correlation, annot=True, cmap='Blues')
sns.heatmap(correlation)
plt.title('HeatMap(graph)')
plt.show()
```

```python
#Question 4:
medication_group = [110, 115, 108, 102, 107, 99, 111, 104, 109, 101]
placebo_group = [120, 125, 130, 122, 128, 119, 124, 127, 123, 126]


from scipy.stats import ttest_ind

stats, p_value =ttest_ind(medication_group, placebo_group)
print("T-Statistic:",stats)
print("P-Value:", p_value)
p_mark = 0.05

if p_value < p_mark:
    print("Reject the Null Hypothesis: The medication is effective.")
else:
    print("Reject the Alternate Hypothesis: The medication is not effective.")
```

```python
#Question 5:
ratings = [8, 9, 7, 5, 6, 10, 9, 4, 7, 8,6, 9, 10, 5, 8, 7, 6, 9, 10, 7]
Mean = np.mean(ratings)
print("Mean Rating:",Mean) 
Median = np.median(ratings)
print("Median Rating:",Median)
Mode = stat.mode(ratings)
print("Mode Rating:",Mode)
Std = np.std(ratings)
print("Standard Deviation:",Std)
sns.histplot(ratings,kde = True)
plt.xlabel('Customer Satisfcation Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Customer Stisfaction Ratings')
plt.show()
```
